```python
import numpy as np
import pandas as pd
import mplfinance as mpf
from scipy.stats import jarque_bera
import pytz
#creating a data frame from a parquet file
df = pd.read_parquet("final_set.parquet")
df = df.set_index('date')
#created a second data frame to hold the number of bars for each day
data2 = {
    'Time':[],
    'Bars':[]
}
df2 = pd.DataFrame(data2)

date = f'{df.index[0]}'
bars = 0
i = 0
print(date)
#traverse the data frame to find duplicate days and add a bar for each duplicate
for index, row in df.iterrows():
    
    if(f'{row["ticker"]}'=="MSFT"):
        if(f'{index}')==(date):
            bars += 1
        else:
            new_row_data = {'Time': date, 'Bars': bars}
            df2.loc[len(df2)] = new_row_data
            date = f'{df.index[i]}'
            bars = 1
    i += 1
new_row_data = {'Time': date, 'Bars': bars}
df2.loc[len(df2)] = new_row_data
ones = 0;
for index, row in df2.iterrows():
    if row['Bars'] == 1:
        ones += 1
print(ones)
print(df2)
df2['Time'] = pd.to_datetime(df2['Time'])
jb_statistic, p_value = jarque_bera(df2['Bars'])
print(f'Jarque-Bera statistic for bars: {jb_statistic}, p-value: {p_value}')
if p_value < 0.05:
    print("The data is not normally distributed.")
else:    print("The data is normally distributed.")
jb_statistic2, p_value2 = jarque_bera(df['dollar volume'])
print(f'Jarque-Bera statistic for dollar volume: {jb_statistic2}, p-value: {p_value2}')
if p_value2 < 0.05:
    print("The dollar volume data is not normally distributed.")
else:    print("The dollar volume data is normally distributed.")
```
